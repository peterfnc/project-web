```python
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
url = 'https://rickandmortyapi.com/api/character/?page=7'




response = requests.get(url)
results = response.json()
results
```

```python
flat_data = json_normalize(results)

flat_data.head()
```

```python
characters = json_normalize(flat_data.results[0])

characters.head()
```

```python
characters.info()
```

```python
data_morty = pd.DataFrame()

for page in range(1,26):
    url = 'https://rickandmortyapi.com/api/character/?page='+str(page)
    response = requests.get(url)
    results = response.json()
    flat_data = json_normalize(results)
    characters = json_normalize(flat_data.results[0])
    data_morty = data_morty.append(characters)
```

```python
data_morty.head(100)
```

```python
data_morty.columns
```

```python
data_morty_clean = data_morty.drop(['created','location.url','origin.url','url'], axis = 1)

data_morty_clean.head()
```

```python
data_morty_clean.columns
```

```python
column_order = ['id','name','gender','species','type','status','origin.name','location.name','episode','image']

data_morty_clean = data_morty_clean[column_order]

data_morty_clean.set_index('id', inplace = True)

data_morty_clean.head(50)
```

```python
data_morty_clean.to_excel('output/morty.xlsx')
```
