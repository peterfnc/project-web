```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import urllib.request
from urllib.request import urlopen
import re
```

### Titles

```python
url = 'https://www.playgroundmag.net/'

html = requests.get(url).content

soup = BeautifulSoup(html, 'lxml')

soup
```

```python
lst = soup.find_all('p', {'class': 'title'}) 

lst
```

```python
titles = [title.get_text() for title in lst]

titles
```

```python
playground_string = "\n".join(titles)

print(playground_string)
```

```python
save_file = open('output/playground.txt', 'w')
save_file.write(playground_string)
save_file.close()
```

### URL-s

```python
url = 'https://www.playgroundmag.net/'

html = requests.get(url).content

soup = BeautifulSoup(html, 'lxml')

links = soup.find_all('a', attrs={'href': re.compile("^https://www.playgroundmag.net")})
link_urls = []
for link in links: 
    link_urls.append(link.get('href'))
    
link_urls
```

```python
article_urls = link_urls[1:-2]
article_urls
```

### Loop

```python
article_texts_lst = []
```

```python
for i in article_urls:
    url = i

    html = requests.get(url).content

    soup = BeautifulSoup(html, 'lxml')
    
    content = soup.find_all('div', {'class':'article_detail-body mce-body p402_premium'}) 
    
    content_clean = [cont.get_text() for cont in content]

    text = ''.join(content_clean)
    
    article_texts_lst.append(text)
    

print(article_texts_lst)    
```

```python
article_text_str = "\n##############\n".join(article_texts_lst)

print(article_text_str)
```

```python
save_file = open('output/playground.txt', 'w')
save_file.write(article_text_str)
save_file.close()
```
